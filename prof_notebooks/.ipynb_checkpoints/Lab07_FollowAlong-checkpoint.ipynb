```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
pd.set_option('display.max_columns', 500)
from IPython.display import display

%matplotlib inline
```

```python
df = pd.read_csv('sampled_footballers.csv')
# Drop the aformentioned columns
model_data = df.drop(['ID','club','club_logo','flag', 'nationality','photo','potential', 'birth_date'], axis = 'columns')
model_data.head()
```

## Randomization test 

```python
# Let's look at weight on overall. 
M1 = ols('overall ~ weight_kg', data=model_data).fit()
M1.summary()
```

```python
# Write a little function that shuffels the column 
# of one single variable. We need to be carful here, 
# as the functions get the data frame as an object 
def randomize_column (df,colname):
    df_copy = df.copy()
    var = df_copy[colname].values
    np.random.shuffle(var)
    df_copy[colname]=var
    return df_copy
```

```python
# Test our randomization function 
D=randomize_column(model_data,'overall')
D.head()
```

```python

# Get the effect we see from the data
observed_effect = ols('overall ~ weight_kg',data = model_data).fit().params[1]

# Initialize somewhere to put the resampled effects
randomization_tests = np.zeros(1000)

for i in range(1000):
    # Randomize the overall
    D = randomize_column(model_data,'overall')

    # Estimate the effect of weight now that we have shuffled overall
    shuffled_model = ols('overall ~ weight_kg',data = D).fit()
    # Stick the result in this array
    randomization_tests[i]=shuffled_model.params[1]

# Plots
plt.hist(randomization_tests)
plt.axvline(observed_effect, color = 'red')

p_value = np.mean(np.abs(randomization_tests)>np.abs(observed_effect))
print('Our p value is: ', p_value)
```

## Parameter-based inference (analytical)

```python
# Let's look at weight on overall. Significant?

M1 = ols('overall ~ weight_kg', data=model_data).fit()
M1.summary()
```

```python
# Let's look at height on overall. Significant?

M2 = ols('overall ~ height_cm', data=model_data).fit()
M2.summary()
```

```python
# What happens if we look at both? What conclusion can we draw?

M3 = ols('overall ~ weight_kg + height_cm', data=model_data).fit()
M3.summary()
```

** Conclusions **
What are we concluding for weight and height in the context of the full model? What if we tested the directed hypothesis that heavier and taller players are better?

```python
# Correlation among predictors

model_data[['height_cm','weight_kg']].corr()
```

```python
# Visualizing correlation among predictors
sns.pairplot(model_data[['overall','height_cm','weight_kg']])
```

## Model-based inference
Compare models using Likelihood-ratio statistics. For linear models, the likelihood ratio statistics is the F-test

```python
# We are comparing the full model against the nested model missing height
M3.compare_f_test(M1)
```

```python
# We are comparing the full model against the nested model missing weight
M3.compare_f_test(M2)
```

```python
# You cannot use the Likelihood ratio test to compare non-nested models! 
M1.compare_f_test(M2)
```

But you can compare any model based on AIC! 

```python
print('Weight vs. weight + height: ',M1.aic - M3.aic) 
print('Height vs. weight + height: ',M2.aic - M3.aic) 
print('weight vs. Height: ',M2.aic - M1.aic) 
```

** Conclusions **
What are we concluding based on the model level?
